```
#importando sparksession
from pyspark.sql import SparkSession

#criando um objeto sparksession object e um appName 
sparkSession=SparkSession.builder.appName("atletas").getOrCreate()
```

```
# Fazendo a leitura do arquivo Bikes.csv
atletas = sparkSession.read.option("header", "true").csv("dbfs:/FileStore/shared_uploads/cezargcpdio@gmail.com/Athletes.csv")
```
