**IMPORTANDO OS DADOS**

```python
atletas = pd.read_excel("Athletes.xlsx")
treinadores = pd.read_excel("Coaches.xlsx")
genero = pd.read_excel("EntriesGender.xlsx")
medalhas = pd.read_excel("Medals.xlsx")
times = pd.read_excel("Teams.xlsx")
```

```python
atletas["NOC"].unique()
```

```python
# Número de Atletas por País.

num_atletas = atletas["NOC"].value_counts().reset_index()
num_atletas.set_index('index', inplace=False)
num_atletas = num_atletas.rename(columns={'index': 'País', 'NOC': 'Num_Atletas'})
num_atletas
```

**INDICADOR  - TOP 10 DE PAÍSES COM MAIOR NÚMERO DE ATLETAS** 

```python
# Top 10 de Países com mais Atletas.

top10_atletas_paises = num_atletas[(num_atletas["Num_Atletas"] >= 324)]
top10_atletas_paises
```

**INDICADOR  - CLASSIFICAÇÃO GERAL DO QUADRO DE MEDALHAS**

```python
medalhas.head()
```

```python
# Quadro Geral de Medalhas.

medalhas2 = medalhas.drop(columns=['Rank by Total'])
medalhas2 = medalhas2.rename(columns={'Gold': 'Ouro', 'Silver': 'Prata'})
medalhas2.set_index('Rank', inplace=True)
medalhas2
```

**INDICADOR 6 - TOP 10 ESPORTES COM MAIS ATLETAS**

```python
modalidade = genero.drop(columns=['Female', 'Male'])
modalidade.head()
```

```python
esportes = modalidade.rename(columns={'Discipline': 'Esporte'})
print(esportes.sort_values(by=['Total'], ascending=False))
```

```python
# Top 10 de Esportes com mAior número de Atletas.

top10_esportes = esportes[(esportes["Total"] >= 297)]
top10_esportes
```
